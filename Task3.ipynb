```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
X = iris.data[:, :4]  # we only take the first 4 features.
y = iris.target
```

```python
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

```python
plt.figure(2, figsize=(10, 10))
plt.clf()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    classifier = KMeans(n_clusters =i , init= 'k-means++',random_state=42)
    classifier.fit(X)
    wcss.append(classifier.inertia_)
plt.plot(range(1,11) ,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.annotate('Elbow structure',(2.8,100))
plt.show()
```

## Seems like there are 3 flowers as the elbox structure is at 3

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st kind")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd kind")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd kind")
ax.w_zaxis.set_ticklabels([])
```
